```python
import pandas as pd
import numpy as np 
import altair as alt
```

```python
rcal = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/RCAL.xlsx") 
scal_pc = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/PcJsw.xlsx") 

rcal[["SampleID","Well","RT"]] = rcal[["SampleID","Well","RT"]].astype(str)
scal_pc[["SampleID","Well"]] = scal_pc[["SampleID","Well"]].astype(str)
```

```python
rcal["RQI"] = 0.0314*np.sqrt(rcal["Permeability"]/rcal["Porosity"])
rcal["PHIZ"] = rcal["Porosity"]/(1-rcal["Porosity"])
rcal["FZI"] = rcal["RQI"]/rcal["PHIZ"]
rcal
```

```python
rcal.dtypes
```

```python
merge = scal_pc.merge(rcal,how="left",on=["Well","SampleID"])
```

```python
merge.head()
```

```python
merge.describe()
```

```python
swc = merge[["Sw","SampleID"]].groupby("SampleID").min().reset_index()
df_plot = rcal.merge(swc,how="left",on="SampleID")
df_plot.rename(columns={"Sw":"Swc"},inplace=True)
df_plot
```

```python
df_plot["Swc"] = df_plot["Swc"].fillna(np.float64(1))
```

```python
df_plot
```

```python
brush = alt.selection_interval()

jsw = alt.Chart(merge).mark_line(clip=True).encode(
    alt.X("Sw:Q",scale=alt.Scale(domain=(0,1.1))),
    alt.Y("JSw:Q",scale=alt.Scale(domain=(0,100))),
    color="RT:N",
    tooltip=["SampleID","PcRes"],
    strokeDash='SampleID:N'
).properties(
    height=600,
    width=500,
    title="Sw vs J(Sw)").transform_filter(brush)

sw_por = alt.Chart(df_plot).mark_circle(size=20).encode(
    alt.X("Porosity:Q"),
    alt.Y("Swc:Q"),
    color=alt.condition(brush,"RT:N",alt.value('lightgray')),
    tooltip=["SampleID","Porosity","Swc"]
).add_selection(brush).properties(
    height=250,
    width=200,
    title="Porosity vs Swc")

sw_perm = alt.Chart(df_plot).mark_circle(size=20).encode(
    alt.X("Permeability:Q",scale=alt.Scale(type='log',base=10,domain=(0.01,10000))),
    alt.Y("Swc:Q"),
    color=alt.condition(brush,"RT:N",alt.value('lightgray')),
    tooltip=["SampleID","Permeability","Swc"]
).add_selection(brush).properties(
    height=250,
    width=200,
    title="Permeability vs Swc")

sw_fzi = alt.Chart(df_plot).mark_circle(size=20).encode(
    alt.X("FZI:Q"),
    alt.Y("Swc:Q"),
    color=alt.condition(brush,"RT:N",alt.value('lightgray')),
    tooltip=["SampleID","FZI","Swc"]
).add_selection(brush).properties(
    height=250,
    width=200,
    title="FZI vs Swc")

pp = alt.Chart(df_plot).mark_circle(size=60).encode(
    alt.X("Porosity:Q"),
    alt.Y("Permeability:Q",scale=alt.Scale(type='log',base=10,domain=(0.01,10000))),
    color=alt.condition(brush,"RT:N",alt.value('lightgray')),
    tooltip=["SampleID","Porosity","Permeability"]
).add_selection(brush).properties(
    height=300,
    width=700,
    title="Porosity vs Permeability")

jsw.mark_circle()+jsw|(pp&(sw_por|sw_perm|sw_fzi))
```
